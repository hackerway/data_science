```python
def country_league():
    
    data = []
    
    html = requests.get("http://www.statto.com/football/stats").text    
    soup = BeautifulSoup(html)
    
    td = soup.findAll("td", {"class" : "col"})
    
    for t in td:
        try:
            tup = [t.findAll("span", {"class" : "cn"})[0].text.encode("utf-8"), 
                   t.findAll("td", {"colspan" : "4"})[0].text.encode("utf-8")]
            data.append(tup)
        except:
            pass
        
    return pandas.DataFrame(data, columns=["country", "league"])  

league_df = country_league()
```

```python
def league_ratings(country, league):
    
    data = []
    
    html = requests.get("http://www.statto.com/football/odds/ratings/" + country + "/" + league + ".html").text    
    soup = BeautifulSoup(html)
    
    tr = soup.findAll("tr", {"class" : ["c0", "c1"]})
        
    for t in tr:    
        d = t.findAll("td")
        tup = [d[1].text.replace(" ", "-"), d[2].text]
        
        data.append(tup)
        
    return pandas.DataFrame(data, columns=["team", "current_index"]).set_index(["team"])

ratings_df = league_ratings("england", "premier-league")
```

```python
def team_stats(team):
    
    data = []
    
    html = requests.get("http://www.statto.com/football/teams/" + team).text    
    soup = BeautifulSoup(html)    
    tr = soup.findAll("table", {"id" : "percent"})
    
    for t in tr:
        td = t.findAll("tr", {"class" : "c1"})        
        wins = td[3].findAll("td")                
        draws = td[4].findAll("td")                
        losses = td[5].findAll("td")        
        supremacy = td[18].findAll("td")
                
        tup = [team, wins[2].text.replace("%", ""), wins[3].text.replace("%", ""), draws[2].text.replace("%", ""), 
               draws[3].text.replace("%", ""), losses[2].text.replace("%", ""), losses[3].text.replace("%", ""),
               supremacy[2].text, supremacy[3].text]
        
        data.append(tup)
        
    return pandas.DataFrame(data, columns=["team", "p_home_wins", "p_away_wins", "p_home_draws", "p_away_draws", 
                                           "p_home_losses", "p_away_losses", "home_supremacy", "away_supremacy"])
```

```python
def make_stats_table():
    
    df = pd.DataFrame()

    for team in ratings_df.index.values:
        df = pd.concat([df, team_stats(team.encode("utf-8"))], ignore_index=True)
            
    df = df.reindex(columns=["team", "p_home_wins", "p_away_wins", "p_home_draws", "p_away_draws", "p_home_losses", 
                               "p_away_losses", "home_supremacy", "away_supremacy"])
    
    return df.set_index(["team"])

stats_table = make_stats_table()
```

```python
def get_team_matches(team):
    
    data = []
    
    html = requests.get("http://www.statto.com/football/teams/" + team + "/2013-2014/results").text    
    soup = BeautifulSoup(html)    
    table = soup.findAll("table")
    
    for row in table:
        tr = row.findAll("tr", {"class" : ["rnl","rnw","rnd","c1"]})
        
        for t in tr:
            d = t.findAll("td")
            try:
                tup = [d[1].text.replace(".","-"), team, d[2].text.replace(" ","-"), d[3].text, d[4].text.replace("&nbsp;", "")[:1]]
            except:
                pass
            
            data.append(tup)
    
    df = pandas.DataFrame(data, columns=["date", "team", "opponent", "ground", "actual_result"])
    
    df = df[:38]
    
    return df
```

```python
def creat_matches_df(ratings_df):

    df = pd.DataFrame()

    for team in ratings_df.index.values:
        df = pd.concat([df, get_team_matches(team)], ignore_index=True)
        
    return df.reindex(columns=["date", "team", "opponent", "ground", "actual_result"])

matches_df = creat_matches_df(ratings_df)
```

```python
def simulate_game(stats_table, ratings_df, team, opponent, ground):
    
    win = 0
    draw = 0
    n_sim = 10000
    
    difference = abs(float(ratings_df.current_index.loc[team]) - float(ratings_df.current_index.loc[opponent]))
        
    for i in range(0, n_sim): 
        if ground == "home":              
            w = float(stats_table.p_home_wins.loc[team])/(0.001 + float(stats_table.p_home_wins.loc[team]) + float(stats_table.p_away_wins.loc[opponent]))            
            if np.random.binomial(1, w) == 1:
                win += 1
                
        elif ground == "away":            
            w = float(stats_table.p_away_wins.loc[team])/(0.001 + float(stats_table.p_away_wins.loc[team]) + float(stats_table.p_home_wins.loc[opponent]))
            if np.random.binomial(1, w) == 1:
                win += 1
                
    if win > 5000+difference:
        return "W", win/100.0
    elif win < 5000-difference:
        return "L", win/100.0
    else:
        return "D", win/100.0
```

```python
def results_team(team):    
    
    df_dict = {}    
    sim_result = []
    sim_percent = []

    team_df = matches_df[matches_df.team == team]

    for row in team_df.itertuples(index=False):
        result, percent = simulate_game(stats_table, ratings_df, row[1], row[2], row[3])   
        
        sim_result.append(result)
        sim_percent.append(percent)
    
    team_df["sim_result"] = sim_result
    team_df["percent_win"] = sim_percent 
        
    return team_df
```
