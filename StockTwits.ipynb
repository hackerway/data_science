## 4. Sentiment Analysis.

In the Stock Market, Bullish traders are hopeful or confident that something or someone will be successful while the bearish are  traders who cause or expect falling stock-market prices.

I sought to create a large data set to classify all the Tweets but there was very little data that was labled. This was due to API rate limits.

I therefore made an assumption that "Bullish" sentiments are usually characterised by "positive" words while "Bearish" sentiments are usually characterised by "negative" words.

I assumed that a "users" sentiment is his/her "rating" for that "symbol". Hence, I moved forward to do Sentiment Analysis on the data collected to determine how the "user" was "rating" a particular symbol at that time.

The data was collected from Rotten Tomatoes API. It was assumed that "Fresh" review implied a "Bullish" sentiment and a "Rotten" review implied a "Bearish" sentiment.

```python
bull_bear = pd.read_csv('critics.csv')

bull_bear = bull_bear[~bull_bear.quote.isnull()]
bull_bear = bull_bear[bull_bear.fresh != 'none']
bull_bear = bull_bear[bull_bear.quote.str.len() > 0]
```

```python
def make_xy(bull_bear, vectorizer=None):    
    text = bull_bear.quote    
    X = vectorizer.fit_transform(text).tocsc() #Call 'fit' to build the vocabulary and 'transform' to convert text to a bag of words
    Y = [1 if x == 'fresh' else 0 for x in bull_bear.fresh] #1/0 array. 1 = fresh review, 0 = rotten review
    return X, Y
```

```python
vectorizer = CountVectorizer(min_df=0.01)    

X, Y = make_xy(bull_bear, vectorizer)

xtrain, xtest, ytrain, ytest = train_test_split(X, Y) #Split up the training data.
clf = MultinomialNB(alpha=1.0) #MultinomialNB classifier.
clf.fit(xtrain, ytrain) #Use the training set to train a MultinomialNB classifier.
test_score = clf.score(xtest, ytest)
```

```python
symbol_title = pd.read_csv("symbol_title.csv", index_col=2)
symbol_title = symbol_title[["symbol", "sector", "industry"]]
symbol_title.head()
```

```python
#Fill in the data "sector" and "industry" to the main dataframe.
def fill_dataframe(x):   
    symbol = list(set(x["symbol"].values))[0]
    sector = symbol_title["sector"][symbol_title["symbol"] == symbol].values
    industry = symbol_title["industry"][symbol_title["symbol"] == symbol].values
    x["sector"] = sector[0] if len(sector) == 1 else np.nan
    x["industry"] = industry[0] if len(industry) == 1 else np.nan    
    return x
```

```python
#Calculate the sentiment for each tweet and assume that that is the rating by the user.
df["rating"] = [clf.predict_proba(vectorizer.transform([text]).tocsc())[0][1] for text in df["text"]]

df = df.groupby("symbol").apply(fill_dataframe) #Fill in missing data

df = df.sort("date") #sort date
df = df.drop_duplicates() #drop duplicates
df = df.dropna() #drop NaN

df["index"] = range(len(df)) #create a new index
df = df.set_index("index") #set new index

df.head()
```

## a. Community Detection.

Social network analysis is the mapping and measuring of relationships and flows between people, groups, organizations, computers, URLs, and other connected information/knowledge entities. 

The nodes in the network are the people and groups while the links show relationships or flows between the nodes. SNA provides both a visual and a mathematical analysis of human relationships. 

To understand networks and their participants, we evaluate the location of actors in the network. Measuring the network location is finding the centrality of a node. These measures give us insight into the various roles and groupings in a network - who are the connectors, mavens, leaders, bridges, isolates, where are the clusters and who is in them, who is in the core of the network, and who is on the periphery?

~Social Network Analysis, A Brief Introduction.

```python
small_df = df[df["date"] == "2013-12-05"] #cut dataframe
```

```python
dict1 = {}
dfx = small_df.groupby("tweet_id")

for tweet_id in list(set(small_df["tweet_id"])):
    dict1[tweet_id] = list(dfx.get_group(tweet_id).symbol.values) #create a dictionary of {tweet_id : [symbols]}
```

```python
#Create a Graph
G = nx.Graph()

for key, value in dict1.items():
    for tup in list(itertools.combinations(value, 2)):        
        u = tup[0]
        v = tup[1] 
        if u != v:
            if G.has_edge(u, v) == True:            
                G[u][v]["weight"] += 1
            else:
                G.add_nodes_from([u, v])
                G.add_edge(u, v, weight=1)
                                
nx.write_gexf(G, "symbol_graph.gexf") #save the Graph
```

## c. Node Influence.

The problem of maximizing inﬂuence spread has been widely studied in social networks, because of its tremendous number of
applications in determining critical points in a social network for information dissemination. 

~On Inﬂuential Node Discovery in Dynamic Social Networks.

```python
dict2 = {}
dfx1 = user_symbol_df.groupby("symbol")

H = nx.Graph()

for symbol in df4["symbol"].values:
    dict2[symbol] = list(dfx1.get_group(symbol).name.values)

for symbol in df4["symbol"].values:
    for user1, user2 in combinations(dict2[symbol], 2):
        if H.has_edge(user1, user2):
            H[user1][user2]["weight"] += 1
            H[user1][user2]["difference"] = 1/float(H[user1][user2]["weight"])            
        else:
            H.add_nodes_from([user1, user2])
            H.add_edge(user1, user2, weight=1, difference=1)
            
T = nx.minimum_spanning_tree(H, "weight")

nx.write_gexf(T, "user_symbol_graph.gexf")
```
